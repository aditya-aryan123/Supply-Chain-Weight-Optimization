```python
import pandas as pd
from itertools import product
import numpy as np
import scipy.stats as ss
```

```python
df = pd.read_csv('Data.csv')
df.head()
```

```python
cat_cols = [col for col in df.columns if df[col].dtypes == 'object']
cat_df = df[cat_cols]
cat_df = cat_df.drop(['Ware_house_ID', 'WH_Manager_ID'], axis=1)
```

```python
cat_df.isnull().sum()
```

```python
cat_df = cat_df.dropna(axis=0)
```

```python
cat_var1 = ('Location_type', 'WH_capacity_size', 'zone', 'WH_regional_zone', 'wh_owner_type', 'approved_wh_govt_certificate')
cat_var2 = ('Location_type', 'WH_capacity_size', 'zone', 'WH_regional_zone', 'wh_owner_type', 'approved_wh_govt_certificate')
```

```python
cat_var_prod = list(product(cat_var1,cat_var2, repeat = 1))
cat_var_prod
```

```python
result = []
for i in cat_var_prod:
    if i[0] != i[1]:
        result.append((i[0], i[1], list(ss.chi2_contingency(
                                        pd.crosstab(cat_df[i[0]], cat_df[i[1]])))[1]))
result
```

```python
chi_test_output = pd.DataFrame(result, columns = ['var1', 'var2', 'coeff'])
chi_test_output.pivot(index='var1', columns='var2', values='coeff')
```

```python
def cramers_V(var1,var2) :
    crosstab =np.array(pd.crosstab(var1,var2, rownames=None, colnames=None)) # Cross table building
    stat = ss.chi2_contingency(crosstab)[0] # Keeping of the test statistic of the Chi2 test
    print(ss.chi2_contingency(crosstab))
    print(stat)
    obs = np.sum(crosstab) # Number of observations
    mini = min(crosstab.shape)-1 # Take the minimum value between the columns and the rows of the cross table
    return (stat/(obs*mini))
```

```python
rows= []

for var1 in cat_df:
    col = []
    for var2 in cat_df :
        cramers =cramers_V(cat_df[var1], cat_df[var2]) # Cramer's V test
        col.append(round(cramers,2)) # Keeping of the rounded value of the Cramer's V  
    rows.append(col)
cramers_results = np.array(rows)
df = pd.DataFrame(cramers_results, columns = cat_df.columns, index =cat_df.columns)
```
